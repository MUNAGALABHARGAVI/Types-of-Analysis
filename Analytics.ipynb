Descriptive Analysis for fitness


```python
import statistics
members=["Bhargavi","Phani","Sunny","Vinny","Sneha"]
ages=[20,21,22,15,23]
workout_hrs=[2,12,15,20,9]
calories_burned=[200,300,500,400,100]
print("Fitness center dataset")
for i in range(len(members)):
  print(members[i],ages[i],workout_hrs[i],calories_burned[i])

#average for workout
avg_workout=sum(workout_hrs)/len(workout_hrs)
print("\nAverage workout hours=",avg_workout)


#Highest calories
max_calories=max(calories_burned)
print("\nMaximum calories burned=",max_calories)

#Lowest calories
min_workout=min(workout_hrs)
print("\nMinimum workout hrs=",min_workout)

#Median calories
med_calories=statistics.median(calories_burned)
print("\nMedian calories=",med_calories)

#Standard deviation
st_dev=statistics.stdev(calories_burned)
print("\nStandard deviation=",st_dev)

#Most Active member
index=calories_burned.index(max_calories)
print("\nMost Active member=",members[index])

#Visual represntation
import matplotlib.pyplot as plt
plt.bar(members,calories_burned)
plt.xlabel("Members")
plt.ylabel("calories_burned")
plt.title("Fitness Analysis")
plt.show()


```

```python
#Diagnosis analysis using e-commerce data
customer=["bharu","sai","sunny","sneha","vinny"]
internet_speed=[2,3,4,1,7]
payment_failure=[2,3,0,4,1]
delivery_charge=[20,30,400,300,100]
purchase_completed=[False,True,True,False,True]
print("Customer Dataset")
for i in range(len(customer)):
  print(customer[i],internet_speed[i],payment_failure[i],delivery_charge[i],purchase_completed[i])
#Total abandment
abandoned=purchase_completed.count(False)
print("\n Total cart abandoned",abandoned)
#Internet Dignosis
print("\n customers with slow internet")
for i in range(len(customer)):
  if internet_speed[i]<5:
    print(customer[i],internet_speed[i],"Mbps")
#payment Diagnosis
print("\n customer with payment failures")
for i in range(len(customer)):
  if payment_failure[i]>0:
    print(customer[i],"Failures",payment_failure[i])
#Diagnose dlivery charge issue
print("\n customers facing highh delivery charges")
for i in range(len(customer)):
  if delivery_charge[i]>150:
    print(customer[i],"delivery charge",delivery_charge[i])
# Find root cause using conditions
print("Root cause Analysis")
for i in range(len(customer)):
  if purchase_completed[i]==False:
    if internet_speed[i]<5:
      print(customer[i],"abandoned cart due to slow internet")
    elif payment_failure[i]>0:
       print(customer[i],"abondoned cart due to payment failure")
    elif delivery_charge[i]>150:
      print(customer[i],"abondoned due to high delivery charge")
  else:
      print(customer[i], "abandoned cart due to unknown reason")
#averade speed
avg_speed=sum(internet_speed)/len(internet_speed)
print("AverageInternet speed",avg_speed)
#visualization
import matplotlib.pyplot as plt
plt.bar(customer,delivery_charge)
plt.xlabel("customers")
plt.ylabel("Delivery chargers")
plt.title("Delivery Charge Analysis")
plt.show()




```

Predictive Analysis

```python
#smart water tank using predictive analysis
import pandas as pd

# Dataset
days = [1,2,3,4,5]
water = [120,150,140,160,180]

print("Smart Water Tank Data\n")

# Total water usage
total = sum(water)
print("Total Water Usage =", total)

# Average usage
avg = sum(water) / len(water)
print("Average Water Consumption =", avg)

# Water usage increase
increase_sum = 0
for i in range(len(water)):
    if i > 0:
        diff = water[i] - water[i-1]
        if diff > 0:
            increase_sum += diff
            print("Day", days[i], "increase =", diff)

print("\nTotal Increased Water Usage =", increase_sum)
```

```python
#using dataset for predictive analysis
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("daily_water_consumption.csv")

# Required columns
data = data[["Days", "Water_Consumed_Liters"]]

print("First 15 Rows")
print(data.head(15))

# Total water usage
total = data["Water_Consumed_Liters"].sum()
print("\nTotal Water Usage =", total)

# Average water usage
average = data["Water_Consumed_Liters"].mean()
print("Average Water Usage =", average)

# Prediction
prediction = average + 1
print("Predicted Next Day Usage =", prediction)

# Water shortage
capacity = 1.5
shortage = average - capacity
print("Water Shortage =", shortage)

# Visualization
plt.plot(data["Days"], data["Water_Consumed_Liters"])
plt.xlabel("Days")
plt.ylabel("Water Usage")
plt.title("Water Consumption")
plt.show()
```

Prescriptive Analysis

```python
#using smart hospital management for prescriptive analysis
# load Data
patients = ["bhargavi", "phani", "sunny", "vinny", "sneha"]
age = [65, 45, 70, 30, 55]
condition = ["Critical", "Moderate", "Critical", "Mild", "Moderate"]
oxygen = [85, 92, 78, 96, 88]
beds_available =[1,1,1,1,1]
print("Hospital Dataset loaded \n")
for i in range(len(patients)):
    print(patients[i], age[i], condition[i], oxygen[i], beds_available[i])
#PRIORITY SCORE
print("\nPriority Scores")
scores = []
for i in range(len(patients)):
    score = 0
    # condition
    if condition[i] == "Critical":
        score += 5
    elif condition[i] == "Moderate":
        score += 3
    else:
        score += 1
    # oxygen level
    if oxygen[i] < 90:
        score += 5
    elif oxygen[i] < 95:
        score += 3
    # age factor
    if age[i] > 60:
        score += 2
    scores.append(score)
    print(patients[i], "Score =", score)
# BED ALLOCATION
print("\nBed Allocation")
for i in range(len(patients)):
    if scores[i] >= 8 and beds_available[i] > 0:
        print(patients[i], "-> bed allocted")
    else:
        print(patients[i], "-> waiting list")
# PRESCRIPTIVE ANALYSIS
print("\nPrescriptive Recommendations")
for i in range(len(patients)):
    if scores[i] >= 8:
        print(patients[i], "-> icu bed requirement")
    elif scores[i] >= 5:
        print(patients[i], "-> normal bed")
    else:
        print(patients[i], "-> home care")
```
